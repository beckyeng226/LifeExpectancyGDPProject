```python
# Import of data and libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('all_data.csv')
#Rename Life Expectancy Column
df.rename(columns={'Life expectancy at birth (years)':'life_expectancy'}, inplace=True)
print(df.head())

print(df.Country.unique())
```

```python
#Assign variable names to individual country's Life Expectancy & GDP
chile_life = df.life_expectancy[(df.Country == 'Chile')]
china_life = df.life_expectancy[(df.Country == 'China')]
germany_life = df.life_expectancy[(df.Country == 'Germany')]
mexico_life = df.life_expectancy[(df.Country == 'Mexico')]
usa_life = df.life_expectancy[(df.Country == 'United States of America')]
zimbabwe_life = df.life_expectancy[(df.Country == 'Zimbabwe')]

chile_gdp = df.GDP[(df.Country == 'Chile')]
china_gdp = df.GDP[(df.Country == 'China')]
germany_gdp = df.GDP[(df.Country == 'Germany')]
mexico_gdp = df.GDP[(df.Country == 'Mexico')]
usa_gdp = df.GDP[(df.Country == 'United States of America')]
zimbabwe_gdp = df.GDP[(df.Country == 'Zimbabwe')]
```

```python
#Summary stats for Life Expectancy in years
mean_life_expect_by_country = df.groupby('Country').life_expectancy.mean()
print(mean_life_expect_by_country)
median_life_expect_by_country = df.groupby('Country').life_expectancy.median()
print(median_life_expect_by_country)
```

```python
# Summary stats for GDP in millions
mean_GDP_by_country = df.groupby('Country').GDP.mean()
print(mean_GDP_by_country)
median_GDP_by_country = df.groupby('Country').GDP.median()
print(median_GDP_by_country)
```

```python
#Histograms of each country's Life Expectancy
labels=['Chile', 'China', 'Germany', 'Mexico', 'USA', 'Zimbabwe']
plt.figure(figsize=(7, 3))
plt.hist(chile_life, alpha=0.5)
plt.hist(china_life, alpha=0.5)
plt.hist(germany_life, alpha=0.5)
plt.hist(mexico_life, alpha=0.5)
plt.hist(usa_life, alpha=0.5)
plt.hist(zimbabwe_life, alpha=0.5)
plt.legend(labels)
plt.show()
```

```python
#Histograms of each country's GDP
plt.figure(figsize=(7, 3))
plt.hist(chile_gdp, alpha=0.5)
plt.hist(china_gdp, alpha=0.5)
plt.hist(germany_gdp, alpha=0.5)
plt.hist(mexico_gdp, alpha=0.5)
plt.hist(usa_gdp, alpha=0.5)
plt.hist(zimbabwe_gdp, alpha=0.5)
plt.legend(labels)
plt.show()
```

```python
#Boxplot to compare Life Expectancy by Country
ax1 = sns.boxplot(data = df, x='Country', y='life_expectancy')
ax1.set_xticklabels(labels, rotation=30)
plt.show()
```

```python
#Boxplot to compare GDP by Country
ax2 = sns.boxplot(data = df, x='Country', y='GDP')
ax2.set_xticklabels(labels, rotation=30)
plt.show()
```

Mean life expectancy in 5 of the 6 countries fell in the mid- to upper 70's, while Zimbabwe showed a much lower mean life expectancy of only 50 years and a wider range of ages.  Median life expectancy was similar to mean life expectancy for all countries.  GDP showed more variability across and within countries.  For example, Chile, Mexico, and Zimbabwe had lower GDPs and varied less from year to year, whereas China and the United States had higher and more varied GDPs across years.  

```python
#Scatterplot to examine overall Life Expectancy and GDP
sns.scatterplot(data=df, x='life_expectancy', y='GDP')
plt.xlabel('Life Expectancy at Birth (in years)')
plt.ylabel('GDP')
plt.title('Life Expectancy at Birth by GDP')
plt.show()
```

```python
#Life Expectancy and GDP in Germany
plt.scatter(germany_life, germany_gdp, color='green')
plt.title('Life Expectancy and GDP for Germany')
plt.xlabel('Life Expectancy in Years')
plt.ylabel('GDP in Millions')
plt.show()
```

```python
# Life Expectancy and GDP in Mexico
plt.scatter(mexico_life, mexico_gdp, color='red')
plt.title('Life Expectancy and GDP for Mexico')
plt.xlabel('Life Expectancy in Years')
plt.ylabel('GDP in Millions')
plt.show()
```

```python
# Life Expectancy and GDP in United States of America
plt.scatter(usa_life, usa_gdp, color='purple')
plt.title('Life Expectancy and GDP for USA')
plt.xlabel('Life Expectancy in Years')
plt.ylabel('GDP in Millions')
plt.show()
```

```python
# Life Expectancy and GDP in Zimbabwe
plt.scatter(zimbabwe_life, zimbabwe_gdp, color='brown')
plt.title('Life Expectancy and GDP for Zimbabwe')
plt.xlabel('Life Expectancy in Years')
plt.ylabel('GDP in Millions')
plt.show()
```

Overall, all six countries demonstrated a positive correlation between Life Expectancy and GDP.  As GDP has increased in each country, Life Expectancy at Birth has also shown an increase.  While 5 countries have shown increases in Life Expectancy of approximately 2-4 years, Zimbabwe has shown an increase of about 15 years over the same time frame.
